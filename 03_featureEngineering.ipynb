# Feature Engineering

```python
import pandas as pd
df_cl = pd.read_csv('example_classification.csv')
df_cl.head()
```

```python
output_name_cl = 'conversion'
y = df_cl[output_name_cl]
X = df_cl.copy()
del X[output_name_cl]
```

## OneHotEncoder (categorical to binarization)

```python
from sklearn.preprocessing import OneHotEncoder

object_columns = X.select_dtypes(include=['object']).columns

encoder = OneHotEncoder(sparse_output=False, drop='first')

X_new = X.copy()

X_encoded = pd.DataFrame(encoder.fit_transform(X[object_columns]))
X_encoded.columns = encoder.get_feature_names_out(object_columns)

X_new = X_new.drop(object_columns, axis=1).reset_index()

X_new = pd.concat([X_new, X_encoded], axis=1)

X_new
```

## MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

object_columns = X.select_dtypes(include=['float64']).columns

scaler = MinMaxScaler()
columns = X[object_columns].columns
X_scaled = scaler.fit_transform(X[object_columns])
X_scaled = pd.DataFrame(X_scaled, columns=columns)

X_scaled
```

## Normalizer

```python
from sklearn.preprocessing import Normalizer

object_columns = X.select_dtypes(include=['float64']).columns

transformer = Normalizer().fit(X[object_columns])
columns = X[object_columns].columns
X_normalized = transformer.transform(X[object_columns])
X_normalized = pd.DataFrame(X_normalized, columns=columns)
X_normalized
```

## PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4) #whiten=True for SVM or k-means
pca.fit(X_new)
X = pca.transform(X_new)

X.shape
```
